```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

def average_price_by_bedroom(df):
    # Sao chép dữ liệu để tránh ảnh hưởng bản gốc
    df_tmp = df.copy()
    
    # Ép kiểu dữ liệu sang số (vì trước đó bạn để là string)
    # 'coerce' sẽ biến các giá trị lỗi thành NaN (rỗng)
    df_tmp['bedrooms'] = pd.to_numeric(df_tmp['bedrooms'], errors='coerce')
    df_tmp['price'] = pd.to_numeric(df_tmp['price'], errors='coerce')
    
    # Loại bỏ các dòng bị thiếu dữ liệu ở 2 cột này
    df_tmp = df_tmp.dropna(subset=['bedrooms', 'price'])
    
    # Gom nhóm theo số phòng ngủ và tính giá trung bình
    # .mean() tính trung bình cộng
    result = df_tmp.groupby('bedrooms')['price'].mean().reset_index(name='average_price')
    
    # Làm tròn 2 chữ số thập phân cho đẹp
    result['average_price'] = result['average_price'].round(2)
    
    return result

# Chạy hàm
df_avg_price = average_price_by_bedroom(newdf)
print(df_avg_price)
```

```python
plt.figure(figsize=(12, 6))

# Vẽ biểu đồ cột
sns.barplot(x='bedrooms', y='average_price', data=df_avg_price, palette='magma')

# Hiển thị con số giá trung bình trên đầu mỗi cột
for i, val in enumerate(df_avg_price['average_price']):
    plt.text(i, val, f"{val}", horizontalalignment='center', verticalalignment='bottom', fontweight='bold')

plt.title('Giá bất động sản trung bình theo số lượng phòng ngủ', fontsize=16)
plt.xlabel('Số lượng phòng ngủ')
plt.ylabel('Giá trung bình (Tỷ VNĐ)')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
dftmp = newdf.copy()
dftmp['price'] = pd.to_numeric(dftmp['price'], errors='coerce')
dftmp['square'] = pd.to_numeric(dftmp['square'], errors='coerce')
dftmp['pricePerSquare'] = dftmp['price'] / dftmp['square']

df_district = dftmp.groupby('district')['pricePerSquare'].mean().reset_index()
df_district = df_district.sort_values('pricePerSquare', ascending=False)
df_district['pricePerSquare'] = df_district['pricePerSquare'].round(2)

plt.figure(figsize=(12, 8), dpi=80)
sns.barplot(x='pricePerSquare', y='district', data=df_district, palette='YlOrRd_r')

for i, val in enumerate(df_district['pricePerSquare']):
    plt.text(val, i, f' {val}', va='center', fontsize=10, fontweight=500)

plt.title("Giá bất động sản trung bình trên mỗi m2 theo Quận", fontsize=18, pad=20)
plt.xlabel("Giá trung bình (VND/m2)", fontsize=12)
plt.ylabel("Quận / Huyện", fontsize=12)
plt.tight_layout()
plt.show()
```
