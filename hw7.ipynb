```python
#xay dung giai thuat K mean
class MyKmeans:
    def __init__(self, k = 3, max_iters = 100):
        self.k = k
        self.max_iters = max_iters
        self.centroids = None
        self.clusters = None
    def fit(self, X):
        self.X = X
        self.n_samples, self.n_features = X.shape

        random_indices = np.random.choice(self.n_samples, self.k, replace=False)
        self.centroids = self.X[random_indices]

        for _ in range(self.max_iters):
            self.clusters = self._assign_clusters(self.centroids)
            old_centroids = self.centroids
            self.centroids = self._update_centroids(self.clusters)
            if np.all(old_centroids == self.centroids):
                break
    def _assign_clusters(self, centroids):
        clusters = [[] for _ in range(self.k)]
        for idx, sample in enumerate(self.X):
            # Tính khoảng cách Euclidean từ điểm đến các trọng tâm
            distances = [np.sqrt(np.sum((sample - c)**2)) for c in centroids]
            # Tìm chỉ số của trọng tâm gần nhất
            closest_idx = np.argmin(distances)
            clusters[closest_idx].append(idx)
        return clusters

    def _update_centroids(self, clusters):
        centroids = np.zeros((self.k, self.n_features))
        for idx, cluster_indices in enumerate(clusters):
            # Lấy giá trị trung bình cộng của các điểm trong cụm
            cluster_mean = np.mean(self.X[cluster_indices], axis=0)
            centroids[idx] = cluster_mean
        return centroids
    def predict(self, X):
        labels = []
        for sample in X:
            distances = [np.sqrt(np.sum((sample-c)**2)) for c in self.centroids]
            labels.append(np.argmin(distances))
        return np.array(labels)
                
```

```python
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt

# Tạo dữ liệu ảo
X, y = make_blobs(n_samples=300, centers=3, cluster_std=0.6, random_state=0)

# Chạy thuật toán tự xây dựng
km = MyKmeans(k=3, max_iters = 100)
km.fit(X)
labels = km.predict(X)

# Vẽ biểu đồ kết quả
plt.scatter(X[:, 0], X[:, 1], c=labels, s=50, cmap='viridis')
plt.scatter(km.centroids[:, 0], km.centroids[:, 1], c='red', s=200, alpha=0.5, marker='X')
plt.title("Kết quả phân cụm K-Means tự xây dựng")
plt.show()
```
